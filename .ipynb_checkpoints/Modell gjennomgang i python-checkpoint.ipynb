```python
# Klassifikasjon som er brukt
classification = file.by_type("IfcClassification")[0]
classification.get_info()
```

```python
# Referanser brukt
classification.HasReferences[0].ClassificationRefForObjects[0].RelatedObjects
```

```python
for reference in classification.HasReferences:
    info = reference.get_info()
    print(" Identification: {}\n Name: {}\n Location: {}".format(
        info["Identification"],info["Name"],info["Location"]))
    print(" Klassifiserer objekter: \n")    
    for obj in reference.ClassificationRefForObjects[0].RelatedObjects:
        print("\tObjekt: {}\n".format(obj))
    print("\n")
```
